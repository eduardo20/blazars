```
def draw_umap(datos_t,n_neighbors=15, min_dist=0.1, n_components=2, 
    metric='euclidean', title='',a=1.576943460405378, alpha=1.0, angular_rp_forest=False,
    b=0.8950608781227859, bandwidth=1.0, gamma=1.0, init='spectral',
    local_connectivity=1.0, metric_kwds={},
    n_epochs=None, negative_sample_rate=5, random_state=42, set_op_mix_ratio=1.0,
    spread=1.0, target_metric='categorical', target_metric_kwds={},
    transform_queue_size=4.0, transform_seed=42, verbose=False):
    import umap
    import matplotlib.pyplot as plt
    datos=datos_t.copy()
    datos['source_type']=2
    fit = umap.UMAP(
        n_neighbors=n_neighbors,
        min_dist=min_dist,
        n_components=n_components,
        metric=metric
    )

    u = fit.fit_transform(datos);
    fig = plt.figure()
    if n_components == 1:
        ax = fig.add_subplot(111)
        ax.scatter(u[:,0], range(len(u)), c=datos_t.spectrum_type )
    if n_components == 2:
        ax = fig.add_subplot(111)
        ax.scatter(u[:,0], u[:,1], c=datos_t.spectrum_type)
    if n_components == 3:
        fig = plt.figure(figsize=(12,7))
        ax = fig.add_subplot(111, projection='3d')
        ax.scatter(u[:,0], u[:,1], u[:,2], s=100,c=datos_t.spectrum_type)
    plt.title(title, fontsize=18)

    
```
